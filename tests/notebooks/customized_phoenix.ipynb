# Customized OTC option

1. dcn: phoenix option without ki barrier, but has coupon barrier

2. fcn: phoenix option without coupon barrier with ki observation only on last day



```python
import datetime
import pyoptmc as opt
calendar = opt.Calendar()

start_date = datetime.date(2025, 11, 5)
ko_ob_dates = calendar.periodic(start_date, '1M', 13, "next")[1:]

mc = opt.MonteCarlo(100, 1000000)
bs = opt.BlackScholes(0.03, 0, 0.265, 244)
end_date = ko_ob_dates[-1]
```

```python
dcn = opt.PhoenixProd(
    start_date= start_date,
    end_date = end_date,
    initial_price = 100.0,
    settlement_barrier = 80.0,
    settlement_dates = ko_ob_dates,
    settlement_coupon_rate = 0.15,
    ko_barrier = 100.0,
    ko_ob_dates = ko_ob_dates,
    ki_barrier = 80.0,
    ki_ob_dates = "daily",
    calendar = calendar)

print(dcn.value(start_date, 100.0, True, mc, bs, request_greeks=True))
```

```python
fcn = opt.PhoenixProd(
    start_date=start_date,
    end_date=end_date,
    initial_price=100.0,
    settlement_barrier=0.0,
    settlement_dates=ko_ob_dates,
    settlement_coupon_rate=0.15,
    ko_barrier=100.0,
    ko_ob_dates=ko_ob_dates,
    ki_barrier=80.0,
    ki_ob_dates=[end_date],
    calendar=calendar)

print(fcn.value(start_date, 100.0, False, mc, bs, request_greeks=True))
```

```python
standard_phx = opt.PhoenixProd(
    start_date= start_date,
    end_date = end_date,
    initial_price = 100.0,
    settlement_barrier = 80.0,
    settlement_dates = ko_ob_dates,
    settlement_coupon_rate = 0.15,
    ko_barrier = 100.0,
    ko_ob_dates = ko_ob_dates,
    ki_barrier = 80.0,
    ki_ob_dates = "daily",
    calendar = calendar)

print(standard_phx.value(start_date, 100.0, False, mc, bs, request_greeks=True))
```
